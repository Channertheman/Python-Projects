```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from statsmodels.formula.api import ols
import statsmodels.api as sm
import seaborn as sns
```

```python
df = pd.read_csv('car_data.csv')
df
```

```python
df
df.isnull().sum();
#no nulls
df['Car_Name'] = df['Car_Name'].str.capitalize()
#capitalize car names for consistency
df['Kms_Driven']=df['Kms_Driven'].divide(1000)
df['Kms_Driven']=df['Kms_Driven'].round(2)
#Divided Kms_Driven column data by 1000, easier to read and compare.
#run this cell once
```

```python
df['Price_Diff']=df['Present_Price']-df['Selling_Price'] 
```

Price difference column. Can build models around this to see what influences price difference.

```python
for i in df['Year']:
    j = i - 2000
    if i > 2000:
        df=df.replace(i,j)
```

Function change year values from 20XX, to XX.
Easier to compare values this way, numbers will be scaled more closely together

```python
#start of exploratory analysis
plt.scatter(x=df['Selling_Price'], y=df['Present_Price'])
```

Plots selling price and present price.
Looks like selling price increases as present price increases, to be expected.
Makes sense because sellers will increase the price of newer models of cars.

```python
plt.scatter(x=df['Year'], y=df['Present_Price'])
```

Plots year and present price. This is pretty flat and I believe it won't affect the regression model too much, but we won't disregard the results.

```python
plt.scatter(x=df['Kms_Driven'], y=df['Present_Price'])
```

Plots kilometers driven and present price.
Relationship here would probably be really weak with all the extreme data points and a heavy concentration of points nearing zero.
Won't rule it out, can test later on.

```python
sns.scatterplot(x=df['Selling_Price'], y=df['Present_Price'], hue=df['Transmission'])
```

Plot of present price, selling price and transmission type
Not extremely conclusive, but it does show that auto transmissions typically sell for more. Can be tested in the regression to determine relationship to selling price.

```python
sns.scatterplot(x=df['Selling_Price'], y=df['Present_Price'], hue=df['Seller_Type'])
```

Plot of present price, selling price and seller.
A lot more conclusive than the previous plot. Dealers sell cars at significantly higher prices than individuals. Worth exploring.

```python
plt.scatter(x=df['Year'], y=df['Price_Diff'])
```

Hypothesis: The Present_Price will be influenced by Present_Price, and Year will have some influence as well. I believe that independent sellers and dealers will have their prices influenced strongly by previous prices of the same models of cars, and year will influence the price because newer models of cars usually will increase in price slowly. The larger the gap in years, the more expensive a car will become.

Secondly, the influence on Price_Diff will be driven strongly by Present_Price and Year. I think Year will be a much stronger driver here than it will be with my first statement above. I believe this will be the case because cars will likely increase in price year after year, so the larger the age gap, the larger the price gap should be as well.

```python
results = ols('Present_Price ~ Selling_Price', data=df).fit()
```

```python
results.summary()
```

Pretty good results. As predicted, selling price and present price create a moderately strong regression model. There is still room for improvement though.
R-squared is quite high meaning it can explain outcomes with decent accuracy. Std err on the variables is low, indicating high accuracy of the data used. With a low p value as well, we can assume our data correlates strongly. Finally, the prob(F statistic) is extremely low, indicating that the model is statistically sifnificant.

```python
results2 = ols('Present_Price ~ Selling_Price + Year', data=df).fit()
```

```python
results2.summary()
```

Year has been added to the new regression model and it has improved by quite a bit. Std error, p values, and Prob(F-statistic) all lie within acceptable ranges while the R-squared saw a significant increase meaning our model is more accurate. The t value for year is lower than we like, indicating it might not be an extremely accurate variable to use in this model.

```python
results3 = ols('Present_Price ~ Selling_Price + Year + Kms_Driven', data=df).fit()
```

```python
results3.summary()
```

Kms_Driven added to the model. As predicted in our analysis, Kms_Driven is not a driver in price. It barely affected the R-squared value, and it's p value is above 0.05 meaning the null is not rejected on too many occasions, skewing our data and weakening the model

```python
results4 = ols('Price_Diff ~ Present_Price', data=df).fit()
```

```python
results4.summary()
```

```python
results5 = ols('Price_Diff ~ Present_Price + Year', data=df).fit()
```

```python
results5.summary()
```

After creating the model for Price_Diff, we see that Present_Price is a strong driver in price difference, with Year adding a significant improvement as well. The t value is still low for year in this model.

```python
results6 = ols('Price_Diff ~ Present_Price + Year + Seller_Type', data=df).fit()
```

```python
results6.summary()
```

For the last model, I wanted to see how Seller_Type affected the model. It provided a minor improvement to the R-squared value, but all it's other results lie within normal limits. Overall, I think it makes the model more accurate but not to a large degree.
