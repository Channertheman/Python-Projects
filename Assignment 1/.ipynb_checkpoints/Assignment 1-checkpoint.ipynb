```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df=pd.read_csv('cereal.csv')
df.isnull().sum();
df;
#no null values, noticed some are negative at a glance
```

```python
df.dtypes;
df2=df[['calories', 'protein', 'fat', 'sodium', 'fiber', 'carbo', 'sugars', 'potass', 'vitamins', 'shelf', 'weight','cups', 'rating']]
columns = df2.columns   
df3=df
df3.columns=df3.columns.str.capitalize()
#creating df3 as the final dataframe, df and df2 used to modify df3. Capitalized column names.
```

```python
avg = 0
for i in columns:
    avg=int(df2[i].mean())
    for j in df2[i]:
        if  j < 0:
            df3=df.replace(j,avg)
#function to replace negative (incorrect) values with more meaningful data

df3['Rating'].unique();
#too many decimal points
```

```python
df3['Rating']=df3['Rating'].round(1)
```

```python
df3['Mfr'].unique();
#used to check unique values
df3['Fat'].mean();
#check average for numerics
```

```python
df4 = df3[['Mfr','Rating']]
y = df4.groupby('Mfr').mean().round(1)
y.plot(kind='bar', figsize=(10,6), title='Average Manufacturer Rating')
#Show average rating per manufacturer. Now why are some preferred over others?
```

```python
df5 = df3.sort_values('Rating', ascending=False).head(20)
df5
#shows top twenty cereals by rating, what makes these so highly rated?
#Low sugar, fat seems to be a trend, also names with wheat or bran in them
```

```python
df6 = df5[['Name', 'Fat', 'Sugars']]
df6.set_index('Name', inplace=True)
df6.plot(kind='bar', figsize=(15,10), title='Fats, Sugars in the Top Twenty Cereals')
#visual of sugars and fats rising with lower ratings
#shows that the top twenty cereals have relatively low levels of sugar and very low levels of fat. most of the top ten have none of both
```

```python
#Now that we've identified a factor in rating, we can move back to the success of brands
#From our previous Mfr graph, we saw that N had the highest average rating. Let's see where its' cereals sits with sugar and fat levels
df7 = df3[['Name','Mfr', 'Fat', 'Sugars']]
z = df7.loc[df7['Mfr']=='N']
z.set_index('Name', inplace=True)
z.plot(kind='bar', figsize=(15,10), title = "N's Fat and Sugar Amounts")
SN = z['Sugars'].mean()
FN = z['Fat'].mean()
print(str(round(SN,2)) + ' average Sugars')
print(str(round(FN,2)) +' average Fats')
#we see here that there are low levels of fat and sugars, some even have none. Let's compare it to two other brands
```

```python
df8 = df3[['Name','Mfr', 'Fat', 'Sugars']]
a = df8.loc[df7['Mfr']=='G']
a.set_index('Name', inplace=True)
a.plot(kind='bar', figsize=(15,10), title = "G's Fat and Sugar Amounts")
SG = a['Sugars'].mean()
FG = a['Fat'].mean()
print(str(round(SG,2)) + ' average Sugars')
print(str(round(FG,2)) +' average Fats')
#This is for the lowest rated brand (G). We see that it has very high levels of sugar and higher amounts of fat when compared to N.
```

```python
df9 = df3[['Name','Mfr', 'Fat', 'Sugars']]
b = df9.loc[df7['Mfr']=='K']
b.set_index('Name', inplace=True)
b.plot(kind='bar', figsize=(15, 10), title = "K's Fat and Sugar Amounts")
SK = b['Sugars'].mean()
FK = b['Fat'].mean()
print(str(round(SK,2)) + ' average Sugars')
print(str(round(FK,2)) +' average Fats')
#K is a middle of the bunch brand. We see that it slightly lower levels of sugar, but less than half of the fats found in G
```
