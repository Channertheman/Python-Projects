```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from statsmodels.formula.api import ols
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('players_stats.csv')
df
```

```python
df.isnull().sum()
```

```python
columns = df.columns.tolist()
```

```python
columns;
```

```python
df2=df.drop(columns=['Birth_Place',
 'Age',
 'Birthdate',
 'Collage',
 'Height',
 'Pos',
 'Age',
 'Team',
 'Weight',
 'BMI'])
```

```python
df2
```

```python
df2['MINS_Per_Game'] = df2['MIN']/df2['Games Played']
df2['PTS_Per_Game'] = df2['PTS']/df2['Games Played']
df2['MINS_Per_Game'] = df2['MINS_Per_Game'].round(1)
df2['PTS_Per_Game'] = df2['PTS_Per_Game'].round(1)
```

```python
df2
```

```python
df2.isnull().sum()
```

```python
#df2['Experience']=df2['Experience'].fillna(value = 'U')
df3=df2.dropna()
df3 = df3.replace('R','0')
df3['Experience'] = df3['Experience'].astype('int64')
x = df3['Experience'].mean()
y = round(x)
```

```python
df2['Experience']=df2['Experience'].fillna(value = str(y))
df2 = df2.replace('R','0')
df2['Experience'] = df2['Experience'].astype('int64')
df2=df2.rename(columns= {'Games Played':'Games_Played', 'FG%': 'FGP', '3P%' : '3PP'})
```

My approach to preparing the dataset was to remove all columns that don't relate to game performance using the .tolist() function and the .drop() function. Fortunately this removed most of the null values issue as well. I did keep the Experience column, because experience in a sport can definetly affect performance. To deal with the null value issue, I dropped all the NaNs and created a new dataframe with the output. I then converted the datatype to int and found the mean. I rounded it, and applied it all the NaNs in the df2. Next I added two new columns: MINS per game and PTS per game so we can get a per game analysis. This was done using the .mean() function and the .round() to make the data more readable.

For analysis, I would like to identify drivers that influence the total amount of points scored in a season, but I want use variables that are not point driven (FGM, FGA, 3PA, etc.). To do this I will use variables like Games_Played and Experience. I also want to use the average columns (PTS_Per_Game, MINS_Per_Game).

Now that we've tested a few variables, we can create our hypotheses.
PTS will increase as more games and minutes are played. Secondly, the higher the average Pts per game, the higher a player's point total will be at the end of the season.

H1: As more games and minutes are played, point total will increase

H1: More points per game will equal higher season point totals

H0: As more games and minutes are played, point total will decrease

H0: More points per game will not equal higher season point totals

```python
results = ols('PTS ~ Games_Played', data=df2).fit()
```

```python
results.summary()
```

```python
results = ols('PTS ~ Games_Played + MINS_Per_Game', data=df2).fit()
```

```python
results.summary()
```

```python
results = ols('PTS ~ Games_Played + MINS_Per_Game + Experience', data=df2).fit()
```

```python
results.summary()
```

```python
results = ols('PTS ~ MINS_Per_Game + PTS_Per_Game + Games_Played', data=df2).fit()
```

```python
results.summary()
```

Now that we have our model, we can explain more of what's going on and how closely related the variables are to point production. 

Overall, we have a strong model. The R-Squared value is quite high, so this model will effective at explaining outcomes. Our Prob (F-Statistic is extremely small, meaning it is statistically significant. The P values are all at zero, so we can safely assume that the null hypothesis is rejected. The T values look good as well, with only Mins_Per_Game being below zero. The std err on some of our variables is higher than we would like, but all the other outcomes are pretty conclusive so we can move forward with this model.

```python
Y = df2['PTS']
X = df2[['Games_Played', 'PTS_Per_Game', 'MINS_Per_Game']]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)
```

```python
train = sklearn.linear_model.LinearRegression().fit(X_train, Y_train)
```

```python
pred = train.predict(X_test)
```

```python
test_rmse = (np.sqrt(mean_squared_error(Y_test, pred)))
test_r2 = r2_score(Y_test, pred)
```

```python
print(test_rmse)
print(test_r2)
```

```python
df3 = pd.DataFrame({'Actual': Y_test, 'Predicted': pred})  
df3
```

Overall, our model is fairly accurate. Our R2 value is quite close to 1, implying that our model is accurate. On the other hand the RMSE value is a little high, telling us that the average difference between predicted and actual values is around 95. I would say that this data is reliable, but I think we can do a bit better with selecting our dependent variables for analysis, which is what I did below.
